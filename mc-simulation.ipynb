## Simulations

### Price

```python
def price():
    price = np.round(
        np.random.normal(
            loc=np.mean(df.iloc[1,1:]),
            scale=np.std(df.iloc[1,1:]),
            size=1),
        0)
    price = int(np.mean(price))
    return price
price()
```

### Fixed Costs

```python
def fixed_costs():
    fc = int(np.round(np.random.uniform(low=df.iloc[6,1], high=df.iloc[6,5]),0))
    return fc
fixed_costs()
```

### Quantity

```python
def quantity():
    q = np.round(
        np.random.normal(
            loc=np.mean(df.iloc[0,1:]),
            scale=np.std(df.iloc[0,1:]),
            size=1),
        0)
    q = int(np.mean(q))
    return q
quantity()
```

### Managment Costs

```python
def mgmt_costs():
    mc = int(np.round(np.random.uniform(low=df.iloc[7,1], high=df.iloc[7,5]),0))
    return mc
mgmt_costs()
```

### Revenue, Sales Costs, Net Income

```python
def revenue():
    revenue = quantity() * price()
    return revenue
revenue()

sales_commission = 0.07

def sales_costs():
    sc = int(revenue() * sales_commission)
    return sc

def income():
    inc = revenue() - sales_costs()
    return inc
income()
```

### Interest, Amortization, Installment

```python
def interest():
    interest = int(np.round(np.random.uniform(low=df.iloc[8,1], high=df.iloc[8,5]),0))
    return interest

amortization = df.iloc[9,1]
depreciation = df.iloc[10,1]
```

### Earnings Before Taxes, Taxes and Earnings After Taxes

```python
def ebt():
    ebt = income() - ( interest() + amortization + installment + mgmt_costs())
    return ebt

def taxes():
    if (int(ebt() * 0.27)) < 0:
        return 0
    else:
        return (int(ebt() * 0.27))

def eat():
    eat = ebt() - taxes()
    return eat

installment = df.iloc[15,1]
```

### Net Income

```python
def net_income():
    ni = (eat() + interest() + depreciation) - installment
    return ni

net_income()
```

### Simulation

```python
# q_sim = int(input("Please, enter number of simulations: "))
q_sim = 1000
mc_net_income = [net_income() for element in range(q_sim)]
botton_lim = 35_000_000
upper_lim = 120_000_000

mean = np.mean(mc_net_income)
std_dev = np.std(mc_net_income)

botton_Z = (botton_lim-mean)/std_dev
upper_Z = (upper_lim-mean)/std_dev

mc_net_income = [net_income() for element in range(q_sim)]

X = round(upper_Z-botton_Z,2)
print(f"{X}%")
print("Mean: ${:,.2f}".format(int(mean)))
print("Std Dev: ${:,.2f}".format(int(std_dev)))
sns.histplot(mc_net_income)
plt.show()
```
